```python
# 
import time  # 
import matplotlib.pyplot as plt  # 
import numpy as np  # 
from scipy.interpolate import make_interp_spline as spline  # 
from scipy.optimize import minimize  # 

# PID
class PID:
    def __init__(self, P=0.2, I=0.0, D=0.0):
        # PID
        self.kp = P  # 
        self.ki = I  # 
        self.kd = D  # 
        self.uPrevious = 0  # 
        self.uCurent = 0  # 
        self.setValue = 0  # 
        self.lastErr = 0  # 
        self.preLastErr = 0  # 
        self.errSum = 0  # 
        self.errSumLimit = 10  # 

    # PID
    def pidPosition(self, curValue):
        # PID
        err = self.setValue - curValue  # 
        dErr = err - self.lastErr  # 
        self.preLastErr = self.lastErr  # 
        self.lastErr = err  # 
        self.errSum += err  # 
        outPID = self.kp * err + (self.ki * self.errSum) + (self.kd * dErr)  # PID
        return outPID

# 
class ComplexNonlinearSystem:
    def __init__(self, initial_temperature=25):
        # 
        self.temperature = initial_temperature

    # 
    def update_temperature(self, voltage):
        # 
        self.temperature = self.temperature + 0.8 * voltage
        return self.temperature

# 
class BeControlled:
    def __init__(self, initial_temperature=25):
        # 
        self.lastControlIn = 0  # 
        self.preLastControlIn = 0  # 
        self.lastControlOut = initial_temperature  # 
        self.preLastControlOut = initial_temperature  # 
        self.complexNonlinearSystem = ComplexNonlinearSystem(initial_temperature)  # 

    # 
    def beControlledDeal(self, outPID):
        # 
        output = self.complexNonlinearSystem.update_temperature(outPID)  # 
        self.preLastControlIn = self.lastControlIn  # 
        self.lastControlIn = outPID  # 
        self.preLastControlOut = self.lastControlOut  # 
        self.lastControlOut = output  # 
        return output

# 
def objective_function(params):
    # 
    P, I, D = params  # 
    pid = PID(P, I, D)  # PID
    beControlled = BeControlled()  # 
    pid.setValue = 1  # 
    curValue = 0  # 
    for i in range(1, 500):  # 
        outPID = pid.pidPosition(curValue)  # PID
        curValue = beControlled.beControlledDeal(outPID)  # 
    final_error = abs(pid.setValue - curValue)  # 
    return final_error

# PID
def testPid(P=0.2, I=0.0, D=0.0, Len=1000):
    # PID
    pid = PID(P, I, D)  # PID
    beControlled = BeControlled()  # 
    pid.setValue = 1  # 1
    curValue = 0  # 0
    # 
    curValueList = []  # 
    timeList = []  # 
    setValueList = []  # 
    PIDoutList = []  # PID
    curValueList.append(0)  # 
    timeList.append(0)  # 
    setValueList.append(pid.setValue)  # 
    PIDoutList.append(0)  # PIDPID

    # PID
    for i in range(1, Len):  # Len
        outPID = pid.pidPosition(curValue)  # PID
        PIDoutList.append(outPID)  # PID
        curValue = beControlled.beControlledDeal(outPID)  # 
        curValueList.append(curValue)  # 
        setValueList.append(pid.setValue)  # 
        timeList.append(i)  # 

    # 
    timeSm = np.array(timeList)  # NumPy
    timeSmooth = np.linspace(timeSm.min(), timeSm.max(), 300)  # 300
    curValueSmooth = spline(timeList, curValueList)(timeSmooth)  # 
    pidoutSmooth = spline(timeList, PIDoutList)(timeSmooth)  # PID

    # 
    plt.figure(0)  # 
    plt.plot(timeSmooth, curValueSmooth, label='Temperature', linewidth=2)  # 
    plt.plot(timeSmooth, pidoutSmooth, label='PID Output', linewidth=2)  # PID
    plt.plot(timeList, setValueList, label='Set Value', linewidth=2)  # 
    plt.xlabel('time (s)')  # x
    plt.ylabel('Value')  # y
    plt.title('PID Control of Nonlinear System')  # 
    plt.ylim((1 - 0.5, 1 + 0.5))  # y
    plt.grid(True)  # 
    plt.legend()  # 
    plt.show()  # 


# PID
initial_guess = [0.3, 0.3, 0.1]  # PID
result = minimize(objective_function, initial_guess, method='SLSQP', bounds=[(0, None), (0, None), (0, None)])  # SLSQP
best_params = result.x  # PID
print(f'Best PID parameters: P={best_params[0]}, I={best_params[1]}, D={best_params[2]}')  # PID

# PID
def test_multiple_pids(params_list, Len=5000):
    plt.figure(figsize=(10, 6))  # 
    for i, params in enumerate(params_list):  # 
        pid = PID(params[0], params[1], params[2])  # PID
        testPid(P=pid.kp, I=pid.ki, D=pid.kd, Len=Len)  # PID
    plt.legend([f'PID {i+1}' for i in range(len(params_list))])  # 
    plt.show()  # 

# 
if __name__ == "__main__":
    params_list = [
        [0.2, 0.1, 0.1],  # PID
        [0.1, 0.2, 0.3],  # PID
        [0.3, 0.2, 0.1],
        [0.15, 0.3, 0.2],
    ]  # PID
    test_multiple_pids(params_list, Len=5000)  # PID

```
